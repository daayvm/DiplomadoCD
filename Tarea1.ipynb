```python
#carga de datos
df = pd.read_csv("/grupos.csv")
```

```python
df.head(5)
```

```python
from scipy import stats
#ALTURA NORMAL?
stat, p_value = stats.shapiro(df['altura'])
print("Shapiro-Wilk test:")
print("Estadístico =", stat)
print("p-value =", round(p_value,4))
print("p-value =", p_value)
```

```python
#ALTURA POR GENERO NORMAL?

altura_M = df[df['genero'] == 'M']['altura']
altura_F = df[df['genero'] == 'F']['altura']

stat_M, p_M = stats.shapiro(altura_M)
stat_F, p_F = stats.shapiro(altura_F)

print("Shapiro-Wilk para género M:")
print("p-value =", round(p_M,6))
print("-------")
print("Shapiro-Wilk para género F:")
print("p-value =", round(p_F,6))
```

```python
from scipy.stats import f_oneway, kruskal
from scipy.stats import shapiro
# Separar alturas por grupo
A = df[df["grupo"] == "A"]["altura"]
B = df[df["grupo"] == "B"]["altura"]
C = df[df["grupo"] == "C"]["altura"]
# Normalidad por grupo
stat, p_A = shapiro(A)
stat, p_B = shapiro(B)
stat, p_C = shapiro(C)

print("Grupo A Shapiro p-value:", round(p_A,5))
print("Grupo B Shapiro p-value:", round(p_B,5))
print("Grupo C Shapiro p-value:", round(p_C,5))

# Interpretación
for g, p in zip(["A","B","C"], [p_A, p_B, p_C]):
    if p > 0.05:
        print(f"Grupo {g}: No se rechaza H0 → datos normales")
    else:
        print(f"Grupo {g}: Se rechaza H0 → datos NO normales")
```

```python
from scipy.stats import kruskal

stat, p_kw = kruskal(A, B, C)
print("Kruskal-Wallis p-value:", round(p_kw,5))

if p_kw > 0.05:
    print("No se rechaza H0 → las alturas son similares entre grupos.")
else:
    print("Se rechaza H0 → hay diferencias significativas entre las alturas de los grupos.")
```

```python
import pandas as pd
from scipy.stats import shapiro, ttest_ind

#carga de datos
df = pd.read_csv("/medias.csv")
df.head(5)
```

```python
prod = df["produccion"]
adm  = df["administracion"]

stat_p, p_prod = shapiro(prod)
stat_a, p_adm = shapiro(adm)

print("Producción Shapiro-Wilk p-value:", round(p_prod, 5))
print("Administración Shapiro-Wilk p-value:", round(p_adm, 5))
```

```python
stat_t, p_ttest = ttest_ind(prod, adm, equal_var=False)  # Welch’s t-test
print("t-test p-value:", round(p_ttest, 5))

alpha = 0.05
if p_ttest > alpha:
    print("No se rechaza H0 → No hay diferencia significativa en la edad promedio entre áreas.")
else:
    print("Se rechaza H0 → Sí hay diferencia significativa en la edad promedio entre áreas.")
```
